```python
#PERCENTAGE OF MISSING VALUES IN EACH COLUMN
(df.isnull().sum()/ len(df) * 100).sort_values(ascending=False)
```

```python
#PLOTTING MISSING VALUES
plt.figure(figsize=(16,6))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
#PLOTTING MISSING VALUES BY PERCENTAGE
plt.figure(figsize=(16,6))
missing_percentage = df.isnull().sum()/len(df) * 100
missing_percentage.plot(kind="bar")
plt.xlabel("Columns")
plt.ylabel("Percentage")
plt.title("Percentage of Missing Values in Each Column")
```

```python
#INSIGHTS FROM DATASET
#WHICH CATEGORY HAS HIGHEST NO OF APPS
df["Category"].value_counts().head(10)
```

```python
#WHICH CATEGORY HAS HIGHEST NUMBER OF INSTALLS
df.groupby("Category")["Installs"].sum().sort_values(ascending=False).head(10)
```

```python
#WHICH CATEGORY HAS HIGHEST NUMBER OF REVIEWS
df.groupby("Category")["Reviews"].sum().sort_values(ascending=False).head(10)
```

```python
#WHICH CATEGORY HAS HIGHEST RATINGS
df.groupby("Category")["Rating"].sum().sort_values(ascending=False).head(10)
```

```python
#RANGE OF RATINGS 
plt.figure(figsize=(16,6))
sns.kdeplot(df["Rating"],color="Blue",shade=True)
```

```python
#WHICH TYPE HAS HIGHEST INSTALLS
plt.figure(figsize=(16,6))
sns.barplot(x="Type",y="Installs",data=df)
```

```python
#WHICH APPS HAS MOST HIGHEST CONTENT RATING
df["Content Rating"].value_counts()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x="Content Rating",y="Installs",data=df)
```

```python
#TOP 5 FREE APPS
plt.figure(figsize=(16,6))
sns.barplot(x="App",y="Rating",data=df[df["Type"]=="Free"].sort_values(by="Rating",ascending=False).head(5))
```

```python
#TOP 5 FREE APPS IN TERMS OF REVIEWS
plt.figure(figsize=(16,6))
sns.barplot(x="App",y="Reviews",data=df[df["Type"]=="Free"].sort_values(by="Reviews",ascending=False).head(5))
```
